```python
import pandas as pd
```

## Load the data and display first 5 and last 5 records

```python
data = pd.read_csv('Prisoners-Dataset.csv')
data.head()
```

```python
data.tail()
```

## Describe the data

```python
data.describe()
```

```python
columns = data.columns.tolist()
columns = columns[2:]
columns
```

## Describe data which has 0 inmates in any column

```python
data.loc[data.isin([0]).any(1)]
```

## Create a column 'total_benefitted' that shows sum of inmates

```python
data['total_benefitted'] = data[columns].sum(axis=1)
data.head()
```

## Create a new row - totals that is the sum of all inmates benefitted

```python
elementary = data['No. of Inmates benefitted by Elementary Education'].sum()
adult = data['No. of Inmates benefitted by Adult Education'].sum()
higher = data['No. of Inmates benefitted by Higher Education'].sum()
computer = data['No. of Inmates benefitted by Computer Course'].sum()
total_benefitted = data['total_benefitted'].sum()
```

```python
list_total = ['All States', 2013, elementary, adult, higher, computer, total_benefitted]
list_total
```

```python
total = pd.DataFrame(data.sum(axis=0))
total
```

```python
import matplotlib.pyplot as plt
```

## Bar char with total benefitted inmates values

```python
plt.figure(figsize=(15, 10))
plt.barh(y=data['STATE/UT'].unique(), width=data['total_benefitted'], data=data)
plt.show()
```

### Andhra Pradesh has the maximum number of beneficiaries.

```python
row_total = pd.DataFrame(data.sum(axis=0))
row_total
```

```python
row_total.drop(['STATE/UT', 'YEAR'], axis=0, inplace=True)
row_total
```

```python
row_total.drop(['total_benefitted'], axis=0, inplace=True)
row_total
```

## Ratio among different modes of benefits

```python
plt.figure(figsize=(30, 20))
plt.pie(row_total, labels=row_total.index.tolist(), autopct='%1.1f%%', textprops=dict(size=20))
plt.show()
```
